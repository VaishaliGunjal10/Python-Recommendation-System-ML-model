```python
fn_user = fn.pivot_table(index='UserID',
                                 columns='BookTitle',
                                 values='BookRating' ) 
fn_user 
```

```python
fn_user.index = fn.UserID.unique()
```

```python
fn_user
```

```python
fn_user.fillna(0, inplace=True)
fn_user
```

```python
ratings
```

```python
ratings.sort_values('num of ratings',ascending=False) 
```

```python
FahrenheitBook=fn_user["Fahrenheit 451"]
StardustBook=fn_user["Stardust"]
FahrenheitBook
```

```python
similar_to_FahrenheitBook = fn_user.corrwith(FahrenheitBook)
similar_to_StardustBook = fn_user.corrwith(StardustBook)
```

```python
corr_FahrenheitBook = pd.DataFrame(similar_to_FahrenheitBook,columns=['Correlation'])
corr_FahrenheitBook.dropna(inplace=True)
corr_FahrenheitBook
```

```python
corr_FahrenheitBook.sort_values('Correlation',ascending=False).head(10)
```

```python
corr_FahrenheitBook = corr_FahrenheitBook.join(ratings['num of ratings'])
corr_FahrenheitBook.head()
```

```python
corr_FahrenheitBook[corr_FahrenheitBook['num of ratings']>3].sort_values('Correlation',ascending=False).head()
```

```python
corr_StardustBook = pd.DataFrame(similar_to_StardustBook,columns=['Correlation'])
corr_StardustBook.dropna(inplace=True)
corr_StardustBook = corr_StardustBook.join(ratings['num of ratings'])
corr_StardustBook[corr_StardustBook['num of ratings']>3].sort_values('Correlation',ascending=False).head()
```
